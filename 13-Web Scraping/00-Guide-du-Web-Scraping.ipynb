## Les composants de base d'un site Web

### HTML
HTML est l'acronyme de `Hypertext Markup Language` et c'est ce que tous les sites Web utilisent pour afficher de l'information. Même les notebooks Jupyter utilisent ce langage pour afficher ces pages dans votre navigateur. Si vous faites un click droit sur une page (et que votre navigateur le supporte, certains doivent activer une option développeur pour cela) vous pourrez visualiser le code HTML de n'importe quelle page Web. 

C'est ce code que Python va accéder et que nous allons analyser.

Jetons un oeil au code HTML d'une page Web simple :

    <!DOCTYPE html>  
    <html>  
        <head>
           <title>Titre de la fenêtre du navigateur</title>
        </head>
        <body>
            <h1>Entête du site Web</h1>
            <p>Un peu de texte dans un paragraphe.</p>
        </body>
    </html>

Voyons ces composants un par un.

Chaque <marqueur> indique un bloc spécifique de la page Web :

    1.<DOCTYPE html> Les documents HTML doivent, devraient, toujours commencer avec cette déclaration qui indique au navigateur que le document est du code HTML 5.
    2. Les blocs composant le document HTML sont placés entre des marqueurs <html> et </html>.
    3. Les meta data et liens vers des fichiers externes de script(comme les fichiers CSS ou JavaScript) sont normalement placés dans le bloc <head>.
    4. Le marqueur <title> donne le titre de la page Web, celui qui s'affiche dans le titre de la fenêtre ou dans l'onglet du navigateur.
    5. Tout ce qui peut être lu par le visiteur du site Web se trouve entre les marqueurs <body> et </body>.
    6. Les titres du texte sont définis dans des blocs qui vont de <h1> à <h6>, où le nombre représente (inversement) la taille du texte ou le niveau de profondeur du titre.
    7. Les paragraphes sont définis par des marqueurs <p>, c'est le texte de base de la page Web.

    Il y a bien plus de marqueurs, comme <a> pour les liens hypertexte, <table> pour les tableaux, <tr> une ligne de tableau et <td> une cellule, et bien d'autres choses !


### CSS

CSS est l'acronyme de `Cascading Style Sheets`, c'est ce qui permet de définir le "style" d'un site Web, dont les couleurs et les polices de caractère, et même certaines animations. CSS utilise des attributs comme **id** ou **class** pour relier un élément HTML précis à une définition CSS, comme une couleur particulière.

  **id** est un identifiant spécifique pour un marqueur HTML qui devra être unique pour tout le document HTML, il devrait donc être utilisé une seule fois sur un seul élément.
  
  **class** permet de définir un style général qui pourra être utilisé sur de nombreux marqueurs HTML.
  
  Pour simplifier, si vous voulez uniquement qu'un élément spécifique soit rouge, vous utiliserez un id, si vous souhaitez que plusieurs marqueurs ou blocs soient rouges, vous créerez une classe dans votre document CSS et vous la relierez à tous ces blocs.

### Guide de scraping

Gardez en tête que vous devez toujours avoir la permission du site Web que vous désirez scraper !
Vérifier les conditions générales d'utilisation et tout texte régulant l'utilisation d'un site avant de vous lancer. Enfin, les sites Web changent régulièrement ! Vous aurez très certainement à mettre à jour votre code régulièrement pour les projets de Web scraping qui durent dans le temps.

On va commencer par une tâche vraiment très simple, récupérer le titre d'une page. Vous vous souvenez sans doute que le marqueur HTML qui le contient est **title**. Pour tester cela nous allons utiliser **www.example.com** qui est un site Web destiné à être utilisé comme domaine d'exemple.

### Project d'application - Manipuler plusieurs pages et items

Nous allons nous attaquer à un projet plus réaliste de scraping d'un site entier. Le site : http://books.toscrape.com/index.html a été précisement créé pour que les apprentis scrapeurs puissent s'entrainer.
Essayons de retrouver le titre de tous les livres qui ont une évaluation à 2 étoiles pour créer une liste Python de tous ces titres.

Voici ce que nous allons faire :

1. Analyser la structure des URL pour naviguer tout le site
2. Récupérer toutes les pages de catalogue
3. Identifier les marqueurs des étoiles d'évaluation et leurs classes
4. Filtrer par évaluation avec un test if 
5. Enregistrer les résultats dans une liste

We can see that the URL structure is the following:

    http://books.toscrape.com/catalogue/page-1.html

```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
```

We can then fill in the page number with .format()

```python
res = requests.get(base_url.format('1'))
```

Now let's grab the products (books) from the get request result:

```python
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
soup.select(".product_pod")
```

Now we can see that each book has the product_pod class. We can select any tag with this class, and then further reduce it by its rating.

```python
products = soup.select(".product_pod")
```

```python
example = products[0]
```

```python
type(example)
```

```python
example.attrs
```

Now by inspecting the site we can see that the class we want is class='star-rating Two' , if you click on this in your browser, you'll notice it displays the space as a . , so that means we want to search for ".star-rating.Two"

```python
list(example.children)
```

```python
example.select('.star-rating.Three')
```

But we are looking for 2 stars, so it looks like we can just check to see if something was returned

```python
example.select('.star-rating.Two')
```

Alternatively, we can just quickly check the text string to see if "star-rating Two" is in it. Either approach is fine (there are also many other alternative approaches!)

Now let's see how we can get the title if we have a 2-star match:

```python
example.select('a')
```

```python
example.select('a')[1]
```

```python
example.select('a')[1]['title']
```

Okay, let's give it a shot by combining all the ideas we've talked about! (this should take about 20-60 seconds to complete running. Be aware a firwall may prevent this script from running. Also if you are getting a no response error, maybe try adding a sleep step with time.sleep(1).

```python
two_star_titles = []

for n in range(1,51):

    scrape_url = base_url.format(n)
    res = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(res.text,"lxml")
    books = soup.select(".product_pod")
    
    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            two_star_titles.append(book.select('a')[1]['title'])
```

```python
two_star_titles
```
