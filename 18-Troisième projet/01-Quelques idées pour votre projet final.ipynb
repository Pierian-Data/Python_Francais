## Algorithmes classiques

**Conjecture de Collatz** - Commencez par un nombre n> 1. Trouvez le nombre de pas qu'il faut pour atteindre un en utilisant le processus suivant: Si n est pair, divisez-le par 2. Si n est impair, multipliez-le par 3 et ajoutez 1.

**Sorting** - Implémentez deux types d'algorithmes de tri: Merge sort et bubble sort.

**Closest pair problem** - Le plus proche paire de points problème ou problème de paire la plus proche est un problème de géométrie computationnelle: donné n points dans l'espace métrique, trouver une paire de points avec la plus petite distance entre eux.

**Sieve d'Eratosthènes** - Le tamis de Eratosthènes est l'un des moyens les plus efficaces pour trouver tous les primes plus petits (moins de 10 millions ou plus).

## Securité

**Chiffement César** - Mettre en œuvre un chiffrement César, codant et décodant. La clé est un entier de 1 à 25. Ce chiffre permet de faire tourner les lettres de l'alphabet (A à Z). L'encodage remplace chaque lettre par la 1ère à la 25ème lettre suivante dans l'alphabet (enveloppe Z à A). Donc la clé 2 chiffre "HI" à "JK", mais la clé 20 crypte "HI" à "BC". Ce simple "chiffre de substitution monoalphabétique" n'offre presque aucune sécurité, car un attaquant qui a le message encodé peut soit utiliser l'analyse de fréquence pour deviner la clé, soit essayer toutes les 25 clés
